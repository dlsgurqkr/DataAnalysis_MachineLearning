```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import pickle
import json
```

```python
df = pd.read_csv('/Users/parkinhyuk/Desktop/GIT/DataAnalysis_MachineLearning/Regression model - KRR/wave.csv')
df.x.head()
```

### Split data

```python
def get_train_test_split(filename: str) -> dict:
    
    # Split the data into training and testing sets with 20% of the data as test
    x_tv, x_test, y_tv, y_test = train_test_split(df.x , df.y, shuffle=True, test_size=0.4)
    
    # Construct the dictionary to return
    split_data = {
        "train": {
            "x": x_tv.values.reshape(-1,1),
            "y": y_tv.values
        },
        "test": {
            "x": x_test.values.reshape(-1,1),
            "y": y_test.values
        }
    }
    
    return split_data
```

```python
def wave(x: np.ndarray) -> np.ndarray:
    return np.exp(-(x)**2)
```

```python
def get_hyper_model() -> GridSearchCV:
    
    parameters = {
        'alpha': [0.1, 0.8, 1, 10, 100],
        'kernel': ['linear', 'rbf', 'poly'],
        'gamma': np.logspace(-4, 1, 6)  # Wide range of gamma values
    }
    
    model = KernelRidge()
    
    hypermodel = GridSearchCV(model,parameters,cv=8,scoring='neg_mean_squared_error')

    
    return hypermodel
```

```python
def train(hyper_model: GridSearchCV, data: dict,name:str) -> KernelRidge:
    
    dt = get_train_test_split(data)
    
    x_tv = dt['train']['x']
    y_tv = dt['train']['y']
    
    x_test = dt['test']['x']
    y_test = dt['test']['y']
    
    #fit model using training data
    hyper_model.fit(x_tv, y_tv)
    
    #Find the best esitmate
    best_model = hyper_model.best_estimator_ 
    
    y_pred_train = best_model.predict(x_tv.reshape(-1,1))
    y_pred_test = best_model.predict(x_test.reshape(-1,1))
    
    
    mae_train = mean_absolute_error(y_tv, y_pred_train)
    mse_train = mean_squared_error(y_tv, y_pred_train)
    r2_train = r2_score(y_tv, y_pred_train)
    
    mae_test = mean_absolute_error(y_test,y_pred_test)
    mse_test = mean_squared_error(y_test,y_pred_test)
    
    r2_test  = r2_score(y_test,y_pred_test)
    

    result = {
        "best_params": {
            "alpha": hyper_model.best_estimator_.alpha,
            "gamma": hyper_model.best_estimator_.gamma,
            "kernel": "rbf"
            
        },
        "train": {
            "mae": mae_train,
            "mse": mse_train,
            "r2":  r2_train
        },
        
        "test": {
            "mae": mae_test,
            "mse": mse_test,
            "r2":  r2_test
        }
    }
    
    with open(f"{name}.json",'w') as file:
        json.dump(result, file, indent=2)
        
    with open(f"{name}.pickl",'wb') as file:
        pickle.dump(hyper_model.best_estimator_, file )
    
    return hyper_model.best_estimator_
```

### Testing Best_Estimoator

```python
#Testing
hyper_model = get_hyper_model()

best_model = train(hyper_model, df, "ex13")
best_model
```

```python
#Plotting
def plot(model: KernelRidge, data: dict) -> mpl.figure.Figure:
    fig = plt.figure()
    
    dt = get_train_test_split(data)
    
    x_tv = dt['train']['x']
    y_tv = dt['train']['y']
    
    x_test = dt['test']['x']
    y_test = dt['test']['y']
    
    x_test = np.array(x_test).reshape(-1, 1)
    
    y_pred_test = model.predict(x_test)
    
    sorted_indices = np.argsort(x_test[:, 0])
    x_test_sorted = x_test[sorted_indices]
    y_test_sorted = y_test[sorted_indices]
    y_pred_test_sorted = y_pred_test[sorted_indices]
    
    # Plot actual vs predicted values
    plt.scatter(x_tv, y_tv, color='blue', label='train')
    plt.scatter(x_test, y_test, color='red', label='test')
    
    plt.plot(x_test_sorted, y_test_sorted, color='green', label='true')
    plt.plot(x_test_sorted, y_pred_test_sorted, linestyle = '--', color='y', label='model')
    
   # plt.plot(x_tv,y_pred_test)
    
    plt.legend()
    plt.xlim(-11,11)
    
  #  plt.title(f"Actual vs Predicted - {name}")
    plt.xlabel("X values")
    plt.ylabel("Y values")

    
    return fig
```

```python
if __name__ == "__main__":
    
    data = get_train_test_split("wave.csv")

    hyper_model = get_hyper_model()

    best_model = train(hyper_model, data, "ex13")
    
    fig = plot(best_model, data) #"ex13"

    plt.show()
```
